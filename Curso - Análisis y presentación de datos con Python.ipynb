## Importación de datos disponibles

En esta parte trabajaremos con pandas y un set de datos 

```python
# Importamos libreria pandas
import pandas as pd
```

```python
missions = pd.read_csv('./data/SpaceMissions.csv')
missions.head()
```

```python
# Accediendo a los datos
# print(missions.loc[1]) # trae toda la 2 fila
print(missions.loc[1,"Company"]) # 2 fila en col "Company"
print(missions.loc[1][0]) # 2 fila en col "Company"
print(type(missions.loc[1][0])) # tipo de dato de un valor
print(missions.loc[1,["Company","Launch Date"]]) # 2 fila en col "Company" y Launch Date
```

```python
# Confiuracion de pandas para visualizacion del dataframe
# Visualizar todas las filas
pd.set_option('display.max_rows',None)
missions
```

```python
# Reestablece la visualizacion de filas
pd.reset_option('display.max_rows')
missions
```
